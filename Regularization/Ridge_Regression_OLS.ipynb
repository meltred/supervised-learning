```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y = make_regression(n_samples=100, n_features=1,n_informative=1,n_targets=1,noise=20,random_state=13)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
from sklearn.linear_model import Ridge # Ridge is nothing but a good Linear Regression 
```

```python
rr = Ridge(alpha=9)
rr.fit(X,y)
print(rr.coef_)
print(rr.intercept_)
```

```python
plt.plot(X,y,'b.')
plt.plot(X,lr.predict(X),color='red',label='alpha=0')
plt.plot(X,rr.predict(X),color='green',label='alpha=10')
plt.legend()
```

```python
class MeraRidge:## My own Ridge Regression
    def __init__(self, alpha=9):
        self.alpha = alpha
        self.m = None
        self.b = None
    def fit(self,X_train,y_train):
        num = 0
        den = 0
        for i in range(X_train.shape[0]):
            num = num + (y_train[i] - y_train.mean()) * (X_train[i] - X_train.mean())
            den = den + (X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean())
        self.m = num /(den + self.alpha)  # Adding alpha to denominator in m to achieve overfitting in Linear Regression
        self.b = y_train.mean() - (self.m * X_train.mean())  
        print(self.m, self.b)
    def predict(X_test):
        return (self.m * X_test) + self.b
```

```python
rr1 = MeraRidge()
rr1.fit(X,y) 
```
